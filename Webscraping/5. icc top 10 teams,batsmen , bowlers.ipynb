```python
#importing the required libraries
from bs4 import BeautifulSoup
import requests
```

```python
page= requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
soup
```

```python
top10_teams = [] # empty list for store the

for i in soup.find_all('span',class_="u-hide-phablet"):
    top10_teams.append(i.text)
    
top10_teams
```

```python
match= [] # empty list for store the

for i in soup.find_all('td',class_="rankings-block__banner--matches"):
    match.append(i.text)
    
match
```

```python
matches= [] # empty list for store the

for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    matches.append(i.text)
    
matches
```

```python
match.append(matches)

print(match)
```

```python
point = [] # empty list for store the

for i in soup.find_all('td',class_="rankings-block__banner--points"):
    point.append(i.text)
    
point
```

```python
points= [] # empty list for store the

for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    points.append(i.text)
    
points
```

```python
rating= [] # empty list for store the

for i in soup.find_all('td',class_="rankings-block__banner--rating u-text-right"):
    rating.append(i.text.replace('\n',' '))
    
rating
```

```python
ratings= [] # empty list for store the

for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    ratings.append(i.text)
    
ratings
```

```python
rating.append(ratings)

print(rating)
```
